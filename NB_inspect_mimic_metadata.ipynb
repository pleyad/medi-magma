```python
import pandas as pd
from pathlib import Path
```

```python
CSV = Path("/srv/scratch1/nbodenmann/mimic-cxr/physionet.org/files/mimic-cxr-jpg/2.0.0/mimic-cxr-2.0.0-metadata.csv")
FILEPATH = Path("/srv/scratch1/nbodenmann/mimic-cxr/physionet.org/files/mimic-cxr-jpg/2.0.0/files/")
```

```python
df = pd.read_csv(CSV)
```

```python
# Identify image bde9f153-11048ce5-af57a7d8-bfce1683-a19befc4
df[df['dicom_id'] == 'bde9f153-11048ce5-af57a7d8-bfce1683-a19befc4']
```

```python
# Show how often different ViewPosition occur
df.sort_values(by=['study_id'], inplace=True)
```

```python
# Iterate over all rows of a patient
for patient_id, patient_images in df.groupby('subject_id'):
    # Iterate over all studies of a patient
    for study_id, study_images in patient_images.groupby('study_id'):
        # Iterate over all images of a study
        AP_images = study_images[study_images['ViewPosition'] == 'AP']
        
        if len(AP_images) > 0:
            image = AP_images.iloc[0]

            
            dicom_id = image['dicom_id']
        
        else:
            PA_images = study_images[study_images['ViewPosition'] == 'PA']
            if len(PA_images) > 0:
                image = PA_images.iloc[0]

                dicom_id = image['dicom_id']

            else:
                print('No frontal images, skipping study', study_id)
                continue
            
        image_path = FILEPATH / f"p{str(patient_id)[:2]}" / f"p{patient_id}" / f"s{study_id}" / f"{dicom_id}.jpg"
        report_path = FILEPATH / f"p{str(patient_id)[:2]}" / f"p{patient_id}" / f"s{study_id}.txt"

        print(image_path)
        print(report_path)
```
